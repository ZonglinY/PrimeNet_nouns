```python
# load super_primitives_may_12.json. Tom manually deleted obscure words from the original version. 
import json
with open("super_primitives_may_12.json", encoding='utf-8') as f:
    super_primitives = json.load(f)
```

```python
# load primenet_may_12.json. All plurals have been converted to singulars and merged in.
with open("primenet_may_12.json", encoding='utf-8') as f:
    primenet = json.load(f)
```

```python
# make a new dictionary that maps entities to primitives; one can use this dictionary to find the primitives that are associated with an entity
keys = list(primenet.keys())
reverse_data = {}
for key in keys:
    for value in primenet[key]:
        if value in reverse_data:
            reverse_data[value].append(key)
        else:
            reverse_data[value] = [key]
```
